# changing column names

```python
data.columns.values
```

```python
#changing the column names so that it can be aesy to handle
data.rename(columns={'iyear':'year','imonth':'month','iday':'day','gname':'group','country_txt':'Country','region_txt':'Region','provstate':'States',
                     'attacktype1_txt':'Attacktype','targtype1_txt':'Targettype','weaptype1_txt':'weapon','nwound':'wound','nkill':'kills',},
                     inplace=True)
```

```python
data.head()
```

# feature selection

```python
data = data[['year','month','day','group','country','Country','Region','States','city','latitude','longitude','target1','summary',
                     'Attacktype','Targettype','weapon','wound','kills','motive']]
```

```python
data.head()
```

```python
data.shape
```

# Handling missing values

```python
data.isnull().sum()
```

```python
data['wound']=data['wound'].fillna(0)
data['kills']=data['kills'].fillna(0)
```

```python
data.info()
```

# Visualisation

```python
#total number of attack in a year
year = data['year'].unique()
years_count = data['year'].value_counts(dropna = False).sort_index()
years_count
```

```python
plt.figure(figsize=(18,10))
sns.barplot(x=year,y=years_count)
plt.xticks(rotation='vertical')
plt.xlabel('Year',fontsize=20)
plt.ylabel('Number of attacks',fontsize=20)
plt.title('Attack in a Year',fontsize=30)
plt.show()
```

```python
#countries with most terror attacks
plt.subplots(figsize=(20,10))
sns.barplot(data['Country'].value_counts()[:10].index,data['Country'].value_counts()[:10].values)
plt.xlabel(['Countries'])
plt.ylabel('Count')
plt.show()
```

```python
# Most number of attack in cities 
data['city'].value_counts().to_frame().sort_values('city',axis=0,ascending=False).head(15).plot(kind='bar',figsize=(20,10))
plt.xlabel('City',fontsize=20)
plt.ylabel('Number of attack',fontsize=20)
plt.show()
```

```python
# Attacktype
data['Attacktype'].value_counts().plot(kind='bar',figsize=(20,10))
plt.xlabel('Type of Attacks',fontsize=20)
plt.ylabel('name of attack',fontsize=20)
plt.title('Kills based on Attacktype',fontsize=30)
plt.show()
```

```python
# Number of people killed in Attacks
data[['Attacktype','kills']].groupby(["Attacktype"],axis=0).sum().plot(kind='bar',figsize=(20,10))
plt.xlabel('Type of attack',fontsize=20)
plt.ylabel('Number of Kills',fontsize=20)
plt.title('Kills based on Attacktype',fontsize=30)
plt.show()
```

```python
# Number of people wounded in different type of Attack
data[['Attacktype','wound']].groupby(["Attacktype"],axis=0).sum().plot(kind='bar',figsize=(20,10),color=['red'])
plt.xlabel('Type of attack',fontsize=20)
plt.ylabel('No. of people wounded',fontsize=20)
plt.title('Wounds based on Attacktype',fontsize=30)
plt.show()
```

```python
# Number of people who are targetted in attack
plt.subplots(figsize=(20,10))
sns.countplot(data['Targettype'],order=data['Targettype'].value_counts().head(10).index)
plt.xlabel('Type of victims',fontsize=10)
plt.ylabel('Count',fontsize=20)
plt.show()
```

```python
# top 10 group attacks
data['group'].value_counts().to_frame().drop('Unknown').head(10).plot(kind='bar',figsize=(20,10))
plt.xlabel('group name',fontsize=20)
plt.ylabel('No. of attacks',fontsize=20)
plt.title('Top 10 Group attacks',fontsize=30)
plt.show()
```

```python
#Country with most number of attacks
df = data[['group','Country','kills']]
countryattack = df.groupby(['group','Country'],axis=0).sum().sort_values('kills',ascending=False).drop('Unknown').reset_index().head()
countryattack
```
